```r
ex9_list = function(df) {
    quant_vars = c("mpg", "displacement", "weight", "acceleration")
    for(var in quant_vars) {
        print(var)
        print(range(df[var]))
        print(mean(df[[var]]))
        print(sd(df[[var]]))
    }
}
```
